```python
import pandas as pd
import numpy as np
data = pd.read_csv('./data/airbnb_hw.csv')
data.head()

#Q2.1
Price_col = data['Price']
print(data.loc[0:50,('Price')],'\n') #examine more data
print(Price_col.unique())
Price_col = Price_col.str.replace(",", "") #replace all commas with empty string (since when over 999 its a string with ,)
Price_col = pd.to_numeric(Price_col,errors='coerce')
print(Price_col.unique())
print(sum(Price_col))

```

```python
#Q2.2
sharks_data = pd.read_csv('./data/sharks.csv')
#print(sharks_data.loc[0:50,('Type')],'\n')
Type_col = sharks_data['Type']
print(Type_col.unique(),'\n') 
print(Type_col.value_counts(), '\n')
Type_col = Type_col.replace(['Boat', 'Boatomg', 'Boating'], 'Watercraft') #all of these can be one name
Type_col = Type_col.replace(['Invalid', 'Unconfirmed','Unverified', 'Under investigation','Questionable'], np.nan) #values that have low counts and aren't good values or proven.
#print(sharks_data['Type'].value_counts())
#need to replace column
sharks_data['Type'] = Type_col
del Type_col
print(sharks_data['Type'].value_counts())
```

```python
#Q2.3
#pretrial_data = pd.read_csv('./data/shark.csv') 
pretrial_data = pd.read_csv('./data/pretrial_data.csv')
pretrial_data.head()

Released_col = pretrial_data['released']
print(Released_col.unique(),'\n') 
print(Released_col.value_counts(), '\n') #get more info on data
Released_col = Released_col.replace(9, np.nan) #for released, has 1 0 or 9, but 9 is a low amount and confusing when representing a true or false
#^ so replace 9s with non numbers
pretrial_data['released'] = Released_col
print(pretrial_data['released'].value_counts()) #shows that 9 was removed
del Released_col
```

```python
#2.4
pretrial_data = pd.read_csv('./data/pretrial_data.csv') #start over so not modified by previous Q
#print(pretrial_data.value_counts(), '\n') #examine data
imposedSentence_col = pretrial_data['ImposedSentenceAllChargeInContactEvent']
# print(imposedSentence_col.unique())
print(pretrial_data['ImposedSentenceAllChargeInContactEvent'].value_counts()) 
sentenceType_col = pretrial_data['SentenceTypeAllChargesAtConvictionInContactEvent']
#print(sentenceType_col.unique())
imposedSentence_col = pd.to_numeric(imposedSentence_col, errors='coerce') #convert from strings to ints
imposedSentence_col = imposedSentence_col.mask(sentenceType_col == 4 , 0) #since by viewing data it shows if it is a type 4 charge it was dismissed normally
imposedSentence_col = imposedSentence_col.mask(sentenceType_col == 9 , 0) #mask here because we don't know what type 9 is
pretrial_data['ImposedSentenceAllChargeInContactEvent'] = imposedSentence_col
print(imposedSentence_col.unique())
del imposedSentence_col
del sentenceType_col
```
